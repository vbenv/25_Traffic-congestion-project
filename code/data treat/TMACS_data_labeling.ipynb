```python
import os
import pandas as pd
import numpy as np

# 변환할 디렉토리 경로
directory = '/Users/gwagseongchan/Desktop/school/플젝&대외활동/25_Traffic-congestion-project/data/TMACS'  # 여기에 디렉토리 경로를 입력하세요

# 새로운 칼럼명
new_columns = [
    "일련번호", "년도", "월", "일", "요일", "시각", "사고등급", "사망자수",
    "중상자수", "경상자수", "부상자수", "연령", "성별", "경력", "음주", "종별",
    "용도", "관련자2", "사고형태", "기상상태", "일광상태", "사고유형1",
    "사고유형2", "법규위반", "도로종류", "ACC_NO", "기초", "지점명"
]

# 디렉토리 내의 모든 .xlsx 파일을 확인
for filename in os.listdir(directory):
    if filename.endswith('.xlsx'):
        file_path = os.path.join(directory, filename)
        
        # 파일 읽기
        df = pd.read_excel(file_path)

        # 첫 두 행 삭제
        df = df.iloc[2:].reset_index(drop=True)

        # 현재 열 수와 새로운 열 수 확인
        current_columns = df.columns.tolist()
        current_length = len(current_columns)
        new_length = len(new_columns)

        # 새로운 칼럼명으로 설정
        df.columns = new_columns[:current_length]  # 현재 열 수에 맞게 새로운 열명 설정

        # 부족한 열 수만큼 빈 열 추가
        if current_length < new_length:
            for col in new_columns[current_length:new_length]:
                df[col] = ''  # 빈 문자열로 추가

        # 수정된 DataFrame을 다시 저장
        df.to_excel(file_path, index=False)
        print(f'수정 완료: {filename}')
```

# 디렉토리 내 파일 xls -> xlsx 변환

```python
import os
import pandas as pd

# 변환할 디렉토리 경로
directory = '/Users/gwagseongchan/Desktop/school/플젝&대외활동/25_Traffic-congestion-project/data/TMACS'  # 여기에 디렉토리 경로를 입력하세요

# 디렉토리 내의 모든 파일을 확인
for filename in os.listdir(directory):
    if filename.endswith('.xls') and filename != '사고누적구간.xls':
        # .xls 파일을 .xlsx로 변환
        file_path = os.path.join(directory, filename)
        xlsx_file_path = os.path.join(directory, filename.replace('.xls', '.xlsx'))
        
        # 파일 읽기
        df = pd.read_excel(file_path)
        
        # .xlsx 파일로 저장
        df.to_excel(xlsx_file_path, index=False)

        print(f'변환 완료: {filename} -> {filename.replace(".xls", ".xlsx")}')
```

# 파일 라벨링 진행하는 코드

```python
import pandas as pd
import os
import re
import openpyxl

def label_accident_files():
    # 사고누적구간 엑셀 파일 읽기
    accumulated_accidents = pd.read_excel('사고누적구간.xls', engine='openpyxl')
    
    # A열("기초")와 B열("지점명") 데이터 가져오기
    base_column = accumulated_accidents.iloc[:, 0]  # 첫 번째 열 (기초)
    site_column = accumulated_accidents.iloc[:, 1]  # 두 번째 열 (지점명)
    
    # 작업 디렉토리 설정
    file_directory = os.getcwd()
    
    # "사고원인조사대상지점-사고구조분석표(상세)" 파일 필터링
    file_pattern = "사고원인조사대상지점-사고구조분석표(상세)"
    matching_files = [f for f in os.listdir(file_directory) if file_pattern in f and f.endswith('.xls')]
    
    # 파일 이름에서 숫자 추출하여 정렬
    def extract_number(filename):
        match = re.search(r'\((\d+)\)', filename)
        return int(match.group(1)) if match else -1
    
    # 파일을 숫자 순서로 정렬
    matching_files.sort(key=extract_number)
    
    # 각 파일에 "기초", "지점명" 열 추가
    for file_name in matching_files:
        try:
            # 파일에서 숫자 추출
            match = re.search(r'\((\d+)\)', file_name)
            if not match:
                print(f"'{file_name}'에서 인덱스를 찾을 수 없습니다.")
                continue
            
            file_index = int(match.group(1))
            
            # 해당 인덱스의 "기초"와 "지점명" 가져오기
            base_value = base_column.iloc[file_index] if file_index < len(base_column) else None
            site_value = site_column.iloc[file_index] if file_index < len(site_column) else None
            
            if base_value is None or site_value is None:
                print(f"'{file_name}' 파일에 매칭되는 '기초' 또는 '지점명' 데이터가 없습니다.")
                continue
            
            # 엑셀 파일 읽기
            file_path = os.path.join(file_directory, file_name)
            df = pd.read_excel(file_path, engine='openpyxl')
            
            # "기초"와 "지점명" 열 추가
            df['기초'] = base_value
            df['지점명'] = site_value
            
            # 파일 저장
            df.to_excel(file_path, index=False, engine='openpyxl')
            print(f"'{file_name}' 파일에 '기초: {base_value}', '지점명: {site_value}' 추가 완료.")
        
        except Exception as e:
            print(f"'{file_name}' 처리 중 오류 발생: {e}")
            import traceback
            traceback.print_exc()

# 스크립트 실행
label_accident_files()
```
